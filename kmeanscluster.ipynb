```python
x=df.iloc[:,3:]
x
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=x['Annual Income (k$)'],y=x['Spending Score (1-100)'])
```

```python
from sklearn.cluster import KMeans

km=KMeans()
```

elbow method

```python
wcss=[]

for i in range(1,10):
  km=KMeans(n_clusters=i,init='k-means++',random_state=1)
  km.fit(x)
  wcss.append(km.inertia_)
```

```python
plt.plot(range(1,10),wcss)
```

```python
from sklearn.cluster import KMeans

km=KMeans(n_clusters=5,init='k-means++',random_state=1)
```

```python
km.fit(x)
```

```python
labels=km.labels_
labels
```

```python
centroid=km.cluster_centers_
centroid
```

```python
y_predict=km.predict(x)
```

```python
y_predict
```

```python
import matplotlib.pyplot as plt

plt.scatter(x=x['Annual Income (k$)'],y=x['Spending Score (1-100)'],c=labels)

plt.scatter(centroid[:,0],centroid[:,1],c='r',marker='*')

plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
print(silhouette_score(x,y_predict))
```
